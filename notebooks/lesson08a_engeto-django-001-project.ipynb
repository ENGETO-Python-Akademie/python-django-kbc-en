### What is Django

---

Imagine you want to build a house. You could start from scratch - making your own bricks, forging your own nails, cutting your own lumber.

Or you could use **prefabricated components** that professionals have already designed and tested.

<br>

**Django** is like a prefabricated house kit for web applications. It provides:

- A **database layer** (ORM) to store and retrieve data
- A **URL routing system** to handle web requests
- A **template engine** to generate HTML
- An **admin interface** out of the box
- **Security features** built-in (CSRF, XSS protection, etc.)
- **Authentication system** ready to use

<br>

### Why Django

---

Django was created in 2003 at a newspaper company (Lawrence Journal-World) where developers needed to build web applications **quickly** with **tight deadlines**.

<br>

| Framework | Philosophy | Best For |
|-----------|------------|----------|
| **Django** | Batteries included | Full-featured web apps, CMS, e-commerce |
| Flask | Minimal, flexible | APIs, microservices, learning |
| FastAPI | Modern, async-first | High-performance APIs |

<br>

**Choose Django when:**
- You need a **full-featured web application** (not just an API)
- You want an **admin interface** without building one
- You prefer **convention over configuration**
- You value **security** and **stability**

<br>

### Batteries Included Philosophy

---

Django's motto is **"The web framework for perfectionists with deadlines."**

This means Django includes everything you need to build a web application:

<br>

| Component | What Django Provides |
|-----------|---------------------|
| Database | ORM with migrations |
| Authentication | User model, login/logout, permissions |
| Admin | Auto-generated admin interface |
| Forms | Form handling and validation |
| Security | CSRF, XSS, SQL injection protection |
| Caching | Multiple cache backends |
| Internationalization | Translation support |

<br>

You don't have to use all of these features, but they're there when you need them.

## Creating a Django Project

---

### Installing Django

---

Before creating a project, you need to install Django. **Always use a virtual environment!**

<br>

**Linux/macOS:**
```bash
# Create and activate virtual environment
python3 -m venv .venv
source .venv/bin/activate

# Install Django
pip install django

# Verify installation
python -m django --version
```

**Windows:**
```powershell
# Create and activate virtual environment
python -m venv .venv
.venv\Scripts\Activate.ps1

# Install Django
pip install django

# Verify installation
python -m django --version
```

<br>

### django-admin startproject

---

Django provides a command-line tool called `django-admin` that helps you create projects and apps.

<br>

To create a new project:

```bash
django-admin startproject mysite
```

<br>

This creates a directory called `mysite` with the following structure:

```
mysite/
├── manage.py
└── mysite/
    ├── __init__.py
    ├── settings.py
    ├── urls.py
    ├── asgi.py
    └── wsgi.py
```

<br>

**Pro tip:** You can also create the project in the current directory:

```bash
# Create project in current directory (note the dot at the end)
django-admin startproject mysite .
```

This avoids the nested directory structure:

```
./                    # Current directory
├── manage.py
└── mysite/
    ├── __init__.py
    ├── settings.py
    └── ...
```

<br>

## Project Structure

---

### Understanding the Generated Files

---

Let's understand what each file does:

<br>

| File | Purpose |
|------|--------|
| `manage.py` | Command-line utility to interact with your project |
| `__init__.py` | Marks the directory as a Python package |
| `settings.py` | Configuration for your Django project |
| `urls.py` | URL declarations - the "table of contents" of your site |
| `asgi.py` | Entry point for ASGI servers (async) |
| `wsgi.py` | Entry point for WSGI servers (traditional) |

<br>

**settings.py** - This is where you configure:
- Database connections
- Installed apps
- Middleware
- Templates
- Static files
- Security settings

````python
# Example: Key settings in settings.py (for illustration)

# SECURITY WARNING: keep the secret key used in production secret!
SECRET_KEY = 'django-insecure-your-secret-key-here'

# SECURITY WARNING: don't run with debug turned on in production!
DEBUG = True

ALLOWED_HOSTS = []  # Add your domain names here in production

# Application definition
INSTALLED_APPS = [
    'django.contrib.admin',      # Admin interface
    'django.contrib.auth',       # Authentication system
    'django.contrib.contenttypes',
    'django.contrib.sessions',   # Session management
    'django.contrib.messages',   # Messaging framework
    'django.contrib.staticfiles', # Static file handling
]
````

<br>

**urls.py** - Maps URLs to views:

````python
# Example: Default urls.py content (for illustration)

from django.contrib import admin
from django.urls import path

urlpatterns = [
    path('admin/', admin.site.urls),  # Admin is available at /admin/
]
````

<br>

### The Inner vs Outer Project Directory

---

A common source of confusion is the **two directories with the same name**:

```
mysite/           # OUTER: Root directory (can be renamed)
├── manage.py
└── mysite/       # INNER: Python package (don't rename!)
    ├── settings.py
    └── ...
```

<br>

| Directory | Can Rename? | Purpose |
|-----------|-------------|--------|
| **Outer** `mysite/` | ✅ Yes | Just a container, can be anything |
| **Inner** `mysite/` | ❌ No | Python package, referenced in settings |

<br>

## manage.py - Your Django Swiss Army Knife

---

The `manage.py` file is your main interface to the Django project. It's essentially a thin wrapper around `django-admin` that:

1. Sets the `DJANGO_SETTINGS_MODULE` environment variable
2. Provides project-specific commands

<br>

**Syntax:**
```bash
python manage.py <command> [options]
```

<br>

### Essential Commands

---

Here are the most important `manage.py` commands you'll use daily:

<br>

| Command | Description |
|---------|-------------|
| `runserver` | Start the development server |
| `migrate` | Apply database migrations |
| `makemigrations` | Create new migrations based on model changes |
| `createsuperuser` | Create an admin user |
| `shell` | Open Python shell with Django context |
| `startapp` | Create a new Django app |
| `check` | Check for common problems |
| `test` | Run tests |

<br>

To see all available commands:

```bash
python manage.py help
```

To get help for a specific command:

```bash
python manage.py help runserver
```

<br>

### Running the Development Server

---

The development server is a lightweight web server built into Django:

```bash
python manage.py runserver
```

<br>

**Output:**
```
Watching for file changes with StatReloader
Performing system checks...

System check identified no issues (0 silenced).

You have 18 unapplied migration(s)...
Run 'python manage.py migrate' to apply them.

February 25, 2026 - 10:00:00
Django version 5.0, using settings 'mysite.settings'
Starting development server at http://127.0.0.1:8000/
Quit the server with CONTROL-C.
```

<br>

**Customizing the server:**

```bash
# Run on a different port
python manage.py runserver 8080

# Run on all interfaces (for network access)
python manage.py runserver 0.0.0.0:8000
```

<br>

**Important:** The development server:
- **Auto-reloads** when you change code
- Is **NOT for production** - use Gunicorn, uWSGI, etc.
- Serves static files automatically (in DEBUG mode)

<br>

## First Steps After Creating a Project

---

After creating your project, here's the typical workflow:

<br>

**1. Apply initial migrations** (creates database tables for built-in apps):

```bash
python manage.py migrate
```

**2. Create a superuser** (admin account):

```bash
python manage.py createsuperuser
```

**3. Run the development server:**

```bash
python manage.py runserver
```

**4. Visit the admin interface:**

Open `http://127.0.0.1:8000/admin/` in your browser and log in!
